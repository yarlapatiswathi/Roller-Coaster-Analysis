# Data Wrangling

```python
df=df[['coaster_name','Location', 'Status','Manufacturer',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean','speed_mph',
       'height_ft','Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['opening_date_clean']=pd.to_datetime(df['opening_date_clean'])
```

```python
df.dtypes
```

```python
df=df.rename(columns={'Location':'location','Status':'status','Manufacturer':'manufacturer','Type_Main':'type_main','opening_date_clean':'opening_date','Inversions_clean':'inversions','Gforce_clean':'gforce'})
```

```python
df.isna().sum()
```

```python
df.loc[df.duplicated()]
# It shows that there are no duplicates
```

```python
# Lets see duplicates by using subset of columns
df[df.duplicated(subset=['coaster_name','location'])].head()

# It seems that there are quite few duplicates to be handled
```

```python
# Handling duplicates
df=df.loc[~df.duplicated(subset=['coaster_name','location','opening_date'])].reset_index(drop=True).copy()
```

```python
df.shape
```

# Answering questions about the data

- what are the locations with the fastest roller coasters?

```python
avg_coaster_speed_loc=df[df['location']!='Other'].groupby('location')['speed_mph'].agg(['mean','count']).query('count>=10').sort_values('mean')
avg_coaster_speed_loc
```

```python
ax=avg_coaster_speed_loc['mean'].plot(kind='barh',figsize=(12,5),title='Average Coaster Speed by Location')
ax.set_xlabel('Average Coaster Speed');
```
